```python
import os

import plotly.express as px
from plotly import graph_objects
import yfinance as yf

import pandas as pd
import numpy as np

import warnings;   warnings.filterwarnings("ignore")
```

#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
%%time
ativos = ['^BVSP', 'ITUB3.SA', 'BBDC4.SA', 'BCSA34.SA']

dt_ini = '2021-06-24'
dados = yf.download(ativos, start=dt_ini)
dados
```

```python
dados.to_csv('./input/dados_bancos_br_3_anos.csv') 
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
for ativo in ativos:
    dados[('Amplitude', ativo)] = dados[('High', ativo)] - dados[('Low', ativo)]
dados.head()
```

```python
fechamento = dados['Amplitude']
fechamento
```

```python
fechamento_vertical = dados['Amplitude'].stack().reset_index()
fechamento_vertical.columns = ['data', 'ativo', 'valor']

fechamento_vertical.head()
```

```python
# Criar pasta output para salvar os gráficos
if not os.path.exists("output"):
    os.makedirs("output")
```

```python
fig = px.line(fechamento_vertical, x='data', y='valor', color='ativo', template='plotly_white')
fig.write_image("output/line_white_fechamento_vertical.png")
fig.show()
```

```python
import os
import plotly.express as px

# Criar a pasta 'output' caso ela não exista
if not os.path.exists("output"):
    os.makedirs("output")

# Gerar o gráfico
fig = px.line(fechamento_vertical, x='data', y='valor', color='ativo', template='plotly_white')

# Salvar o gráfico na pasta 'output' em formato PNG
fig.write_image("output/line_white_fechamento_vertical.png")

# Exibir o gráfico
fig.show()
```

```python
fig = px.histogram(fechamento_vertical, x='data', y= 'valor', color='ativo', template='plotly_dark')
fig.write_image("output/histograma_dark_fechamento_vertical.png")
fig.show()
```
