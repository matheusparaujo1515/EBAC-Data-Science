```python
import pandas as pd

# Carregar o arquivo Feather (.ftr)
df = pd.read_feather('credit_scoring.ftr')

# Remover colunas desnecessárias (exemplo, substitua 'colunas_a_remover' pelas colunas que você quer excluir)
colunas_a_remover = ['coluna1', 'coluna2']  # Exemplo, adapte à sua necessidade
df = df.drop(columns=colunas_a_remover)

# Amostragem dos dados (opcional, use uma fração menor do dataset, por exemplo, 80% dos dados)
df_sample = df.sample(frac=0.8, random_state=42)  # Exemplo de 80% dos dados

# Converter tipos de dados para reduzir espaço (Exemplo: converter floats para float32, int64 para int32)
df_sample['coluna_exemplo'] = df_sample['coluna_exemplo'].astype('float32')

# Salvar o arquivo com compressão no formato Parquet (que suporta compressão)
df_sample.to_parquet('caminho_para_salvar_arquivo/arquivo_reduzido.parquet', compression='gzip')

# Ou salvar novamente como Feather (sem compressão)
df_sample.reset_index(drop=True, inplace=True)
df_sample.to_feather('caminho_para_salvar_arquivo/arquivo_reduzido.ftr')
```
